```python
from tensorflow.contrib.learn.python.learn.datasets.mnist import read_data_sets
import tensorflow as tf
```

```python
mnist = read_data_sets("./dataset", one_hot=True, reshape=False, validation_size=0)
```

```python
class Network(object):
    def __init__(self):
        self.x = tf.placeholder(tf.float32, [None, 28, 28, 1], name = "x_input")
        self.W1 = tf.Variable(tf.truncated_normal([5, 5, 1, 4], stddev=0.1))
        self.B1 = tf.Variable(tf.ones([4])/10)
        self.stride = 1
        self.Y1 = tf.nn.conv2d(self.x, self.W1, strides=[1, self.stride, self.stride, 1], padding='SAME') + self.B1
        self.output = tf.nn.relu(self.Y1, name = 'z_output')
```

```python
import numpy as np
x, label = mnist.train.next_batch(32)
```

```python
class Train(object):
    
    def __init__(self):
        self.network = Network()
        self.sess = tf.Session()
        self.sess.run(tf.global_variables_initializer())
        self.saver = tf.train.Saver(max_to_keep = 0)
        self.CKPT_DIR = 'new_ckpt'
        
    def train(self):
        output = self.sess.run(self.network.output, feed_dict={self.network.x:x})
        self.saver.save(self.sess, self.CKPT_DIR + '/model',global_step = 10000)
        print (output.tolist())
```

```python
if __name__ == '__main__':
    model = Train()
    model.train()
```

```python
import numpy as np
y, label = mnist.train.next_batch(32)
```

```python
def model_restore_inference():
    sess = tf.Session()
    saver = tf.train.import_meta_graph('./new_ckpt/model-10000.meta')
    saver.restore(sess, tf.train.latest_checkpoint('./new_ckpt'))
    input_x = sess.graph.get_tensor_by_name('x_input:0')
    logits = sess.graph.get_tensor_by_name('z_output:0')
    ret = sess.run(logits, {input_x:y})
    print (ret.tolist())
```

```python
model_restore_inference()
```
